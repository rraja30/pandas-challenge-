```python
#mine

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)

student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#count total number of schools
total_schools = len(school_data_complete["school_name"].unique())
total_schools
```

```python
#count total number of students
total_student = len(school_data_complete["Student ID"].unique())
total_student
```

```python
#count total budget
budget= school_data_complete['budget'].unique().sum()
budget
```

```python
#average math score 
avg_math = school_data_complete['math_score'].sum()/len(school_data_complete["Student ID"].unique())
avg_math
```

```python
#average reading score
avg_read = school_data_complete['reading_score'].sum()/len(school_data_complete["Student ID"].unique())
avg_read
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
pass_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
pass_math_perc = pass_math_count / float(total_student) * 100
pass_math_perc
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_read_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
pass_read_perc = pass_read_count / float(total_student) * 100
pass_read_perc
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass = school_data_complete[(school_data_complete["reading_score"]>= 70) & 
                                    (school_data_complete["reading_score"]>= 70)].count()["Student ID"]
overall_pass_rate = overall_pass/ float(total_student)*100
overall_pass_rate
```

```python
# Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Number of Schools": [total_schools], 
                                 "Total Number of Students": total_student, 
                                 "Total Budget": budget, 
                                 "Average Math Score": avg_math, 
                                 "Average Reading Score": avg_read,
                                 "% Passing Math": pass_math_perc,
                                 "% Passing Reading": pass_read_perc, 
                                 "% Overall Passing":overall_pass_rate
                                })

district_summary["Total Number of Students"] = district_summary["Total Number of Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```

```python
# School Name
school_name = school_data_complete["school_name"].unique()
school_name
```

```python
# School Type
school_type = school_data.set_index(["school_name"])["type"]
school_type
```

```python
# Total Students
school_stud_count = school_data_complete["school_name"].value_counts()
school_stud_count
```

```python
# Total School Budget
school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
school_budget
```

```python
# Per Student Budget
per_school_bud = school_budget / school_stud_count
per_school_bud
```

```python
# Average Math Score
indv_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
indv_school_math
```

```python
# Average Reading Score
indv_school_read = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
indv_school_read
```

```python
# % Passing Math
school_pass_math = school_data_complete[(school_data_complete["math_score"] >= 70)].groupby(["school_name"])["math_score"].count()
school_pass_math
```

```python
# % Passing Reading
school_pass_read = school_data_complete[(school_data_complete["reading_score"] >= 70)].groupby(["school_name"])["math_score"].count()
school_pass_read
```

```python
# % Overall Passing (The percentage of students that passed math and reading.)
perc_math_pass = school_pass_math / school_stud_count * 100
perc_read_pass = school_pass_read / school_stud_count * 100
perc_overall_pass = (perc_math_pass + perc_read_pass) / 2

perc_overall_pass
```

```python
# Create a dataframe to hold the above results
indv_school_summary = pd.DataFrame({"School Type": school_type, 
                                    "Total Students": school_stud_count,
                                    "Total School Budget": school_budget,
                                    "Per Student Budget" : per_school_bud, 
                                    "Average Math Score": indv_school_math, 
                                    "Average Reading Score": indv_school_read, 
                                    "% Passing Math": perc_math_pass, 
                                    "% Passing Reading": perc_read_pass,
                                    "% Passing Overall":perc_overall_pass})

# Formatting
indv_school_summary["Total School Budget"] = indv_school_summary["Total School Budget"].map("${:,.2f}".format)
indv_school_summary["Per Student Budget"] = indv_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
indv_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Given bins
spending_bins = [0, 585, 630, 645, 680]
bin_names = ["<$585", "$585-630", "$630-645", "$645-680"]
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = indv_school_summary.copy()
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_bud, spending_bins, labels=bin_names)
#Calculate averages
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Overall"]
#spending_summary dataframe
# Assemble into data frame
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Passing Overall": overall_passing_spending})
spending_summary = spending_summary[["Average Math Score", 
                            "Average Reading Score", 
                            "% Passing Math", "% Passing Reading",
                            "% Passing Overall"]]
spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
size_bin_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
indv_school_summary["School Size"] = pd.cut(indv_school_summary["Total Students"], size_bins, labels=size_bin_names)
#Calculate Averages
size_math_scores = indv_school_summary.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = indv_school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = indv_school_summary.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = indv_school_summary.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = indv_school_summary.groupby(["School Size"]).mean()["% Passing Overall"]
#Creating dataframe
size_summary = pd.DataFrame({"Average Math Score" : size_math_scores,
                                 "Average Reading Score": size_reading_scores,
                                 "% Passing Math": size_passing_math,
                                 "% Passing Reading": size_passing_reading,
                                 "% Passing Overall": size_overall_passing})
size_summary = size_summary[["Average Math Score", 
                            "Average Reading Score", 
                            "% Passing Math", "% Passing Reading",
                            "% Passing Overall"]]
size_summary
```
